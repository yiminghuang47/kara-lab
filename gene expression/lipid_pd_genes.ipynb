```python
lipid_file = "lipid_genes.txt"
all_lipid_gene_list = []
with open(lipid_file,"r") as file:
    for line in file:
        line = line.strip()
        genes = line.split(",")
        for gene in genes:
            gene = gene.strip()
            if gene:
                all_lipid_gene_list.append(gene)

PD_file = "PD_genes.txt"
all_PD_gene_list = []
with open(PD_file,"r") as file:
    for line in file:
        line = line.strip()
        genes = line.split(",")
        for gene in genes:
            gene = gene.strip()
            if gene:
                all_PD_gene_list.append(gene)
print(all_lipid_gene_list)
print(all_PD_gene_list) 

```

```python
import pandas as pd
df = pd.read_excel("lipid_PD_genes.xlsx",header=1)
lipid_genes = []
PD_genes = []
both_genes = []
for gene_list_string in df["Gene"]:
    genes = gene_list_string.split(",")
    lipid_genes_temp = []
    PD_genes_temp = []
    both_genes_temp = []  
    for gene in genes:
        gene = gene.strip()
        if gene in all_lipid_gene_list:
            lipid_genes_temp.append(gene)
        if gene in all_PD_gene_list:
            PD_genes_temp.append(gene)
        if gene in all_lipid_gene_list and gene in all_PD_gene_list:
            both_genes_temp.append(gene)
    lipid_genes.append(','.join(lipid_genes_temp)) 
    PD_genes.append(','.join(PD_genes_temp))
    both_genes.append(','.join( both_genes_temp))
print(lipid_genes)
print(PD_genes)
print(both_genes)
df["Lipid Genes"] = lipid_genes
df["PD Genes"] = PD_genes
df["Both Genes"] = both_genes
# Save the updated DataFrame to a new Excel file
output_file = "lipid_PD_genes_updated.xlsx"
df.to_excel(output_file, index=False)
```

```python
# find common elements in both lists
common_lipid_PD_genes = set(all_lipid_gene_list) & set(all_PD_gene_list)
# output to file
with open("common_lipid_PD_genes.txt", "w") as f:
    for gene in common_lipid_PD_genes:
        f.write(gene + "\n")
```
